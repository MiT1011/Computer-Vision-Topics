```python
#Gray scale Conversion
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load an image
image = cv2.imread("/content/drive/MyDrive/FaceDetectionImage.jpg")

# Convert to grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display original and grayscale images
cv2_imshow(image)
print("\n\nGray scale Image \n")
cv2_imshow(grayscale_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resize image
resized_image = cv2.resize(image, (int(image.shape[1]/3), int(image.shape[0]/3)))

# Display original and resized images
cv2_imshow(resized_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Extract blue channel
blue_channel = image[:, :, 0]

# Display original and blue channel images
print("Blue Channel")
cv2_imshow(blue_channel)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Crop Image
# Define crop coordinates (adjust as needed)
x_start, y_start, width, height = 400, 200, 400, 400

# Crop the image
cropped_image = image[y_start:y_start+height, x_start:x_start+width]

# Display original and cropped images
cv2_imshow(cropped_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Image Rotation
# Define rotation angle in degrees (positive for clockwise)
angle = 180

# Get image center
center = (image.shape[1] // 2, image.shape[0] // 2)

# Create rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Rotate the image
rotated_image = cv2.warpAffine(image, rotation_matrix, image.shape[:2])

# Display original and rotated images
cv2_imshow(rotated_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Flip image horizontally
flipped_horizontal = cv2.flip(image, 1)

# Flip image vertically
flipped_vertical = cv2.flip(image, 0)

# Display original and flipped images
cv2_imshow(flipped_horizontal)
print("\n")
cv2_imshow(flipped_vertical)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
